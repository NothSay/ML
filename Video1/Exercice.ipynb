```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("Exercice.csv")
df
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Capital')
plt.scatter(df.year, df['per capita income (US$)'], color = 'red', marker='+')
```

```python
prices = df["per capita income (US$)"]
years = df.drop('per capita income (US$)',axis='columns')
prices
```

```python
reg = linear_model.LinearRegression()
reg.fit(years,prices)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
828.46507522 * 2020 - 1632210.7578554575
```

```python
df_new = pd.DataFrame({
    'year' : [2017, 2018, 2019, 2020, 2021],
})
df_new
```

```python
d = reg.predict(df_new)
df_new['per capita income (US$)'] = d
df_new
```

```python
dd = df.append(df_new)
dd
```

```python
dd.to_csv('ResouExercice.csv', index=False)
```

```python
%matplotlib inline
plt.xlabel('years')
plt.ylabel('per capita income (US$)')
plt.scatter(dd.year,dd['per capita income (US$)'],color='red',marker='+')
plt.plot(dd.year, reg.predict(dd[['year']]), color='blue')
```
